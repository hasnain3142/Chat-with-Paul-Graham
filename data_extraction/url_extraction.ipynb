```python
from openai import OpenAI
from pymongo import MongoClient
import tiktoken
from langchain.text_splitter import RecursiveCharacterTextSplitter
from tenacity import retry
from tenacity import stop_after_delay
from tenacity import RetryError
from tenacity import stop_after_attempt
from tenacity import wait_exponential
from bs4 import BeautifulSoup
import uuid

mongodb_url = ""
mongodb_database = "paul_graham"
mongodb_collection = "data"

error_urls = []
client_mongo = MongoClient(mongodb_url)
db = client_mongo[mongodb_database] 

embedding_client = OpenAI(api_key="", base_url="https://api.aimlapi.com", timeout=60)

def tiktoken_len(text):
  tokenizer = tiktoken.get_encoding('cl100k_base')
  tokens = tokenizer.encode(text, disallowed_special=())
  return len(tokens)
    
text_splitter = RecursiveCharacterTextSplitter(
  chunk_size=400,
  chunk_overlap=20,
  length_function=tiktoken_len,
  separators=['\n\n', '\n', ' ', ''])


@retry(wait=wait_exponential(multiplier=1, min=4, max=10), stop=stop_after_attempt(10))
def create_embedding(client,text):
    try:
        response = client.embeddings.create(
            model="text-embedding-ada-002",
            input=text
        )
        return (response, 'success')     
    except Exception as e:
        if 'The server is currently overloaded with other requests' in e:
            raise Exception
        else:
            return (e, 'failed')
            

for idx,url in enumerate(list(set(urls))[2:]):
    print("idx: ",idx)
    page_response = requests.get(url)
    if page_response.status_code == 200:
        page_soup = BeautifulSoup(page_response.text, 'html.parser')
        page_text = page_soup.get_text(separator='\n', strip=True)
        print("url: ",url)
        splitted_content = text_splitter.split_text(page_text)        
        splitted_content = [
            ' '.join(i.split()) for i in splitted_content if i.strip() != ''
          ]   
        try:
            splitted_content.remove('.')
        except:
            pass        
        return_json_data = []
        for chunk in splitted_content:
            temp_json = {}
            try:
                embedding_resp = create_embedding(embedding_client,chunk)
                if embedding_resp[1] == 'failed':
                    print("url: ",url)
                    print(embedding_resp[0])
                else:
                    response = embedding_resp[0]
            except RetryError:
                pass        
            temp_json['_id'] = str(uuid.uuid4())
            temp_json['content'] = chunk
            temp_json['values'] = response.data[0].embedding
            temp_json['url'] = url
            return_json_data.append(temp_json)    
        collection = db[mongodb_collection]
        collection.insert_many(return_json_data, ordered=False)       
        print("successful extraction and embedding")        
        print("--------------")
    else:
        error_urls.append(url)
        print("url: ",url)
        print("error occured: ",page_response.status_code)
        print("--------------")
```

```python
from langchain_mongodb import MongoDBAtlasVectorSearch
from langchain_openai import OpenAIEmbeddings

mongodb_url = ""
mongodb_database = "paul_graham"
mongodb_collection = "data"

query = "Who is Paul Graham?"
client_mongo = MongoClient(mongodb_url)
db = client_mongo[mongodb_database] 

client_mongo = MongoClient(mongodb_url)
collection = db[mongodb_collection]
embedding_client = OpenAIEmbeddings(api_key="", timeout=60)

vector_search = MongoDBAtlasVectorSearch(
    collection = collection,
    embedding = embedding_client,
    index_name="vector_index",
    text_key="content",
    embedding_key="values",
    relevance_score_fn="cosine"
)   
client = OpenAI(api_key="", base_url="https://api.aimlapi.com", timeout=60)
response = client.embeddings.create(
            model="text-embedding-ada-002",
            input=query
        )
search_results = vector_search.similarity_search(query=query, k=5, embeddings=response.data[0].embedding)
```
